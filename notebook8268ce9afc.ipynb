```python
import matplotlib.pyplot as plt
import numpy as np

# Basic Line Plot
x = np.linspace(0, 10, 100)  # Create 100 evenly spaced points from 0 to 10
y = np.sin(x)  # Calculate the sine of each x value

plt.plot(x, y)
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("Sine Wave")
plt.show()


# Scatter Plot
x = np.random.rand(50)  # 50 random x values
y = np.random.rand(50)  # 50 random y values

plt.scatter(x, y, c='red', marker='o', label="Data Points")  # Red circles
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("Scatter Plot")
plt.legend()  # Show the legend
plt.show()


# Bar Chart
categories = ['A', 'B', 'C', 'D']
values = [25, 15, 30, 10]

plt.bar(categories, values, color='skyblue')
plt.xlabel("Categories")
plt.ylabel("Values")
plt.title("Bar Chart")
plt.show()


# Histogram
data = np.random.randn(1000)  # 1000 random numbers from a normal distribution

plt.hist(data, bins=30, color='lightgreen', edgecolor='black') # 30 bins
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.title("Histogram")
plt.show()


# Multiple Plots in One Figure (Subplots)
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

plt.figure(figsize=(10, 5))  # Adjust figure size

plt.subplot(1, 2, 1)  # 1 row, 2 columns, first plot
plt.plot(x, y1, color='blue')
plt.title("Sine Wave")

plt.subplot(1, 2, 2)  # 1 row, 2 columns, second plot
plt.plot(x, y2, color='red')
plt.title("Cosine Wave")

plt.tight_layout()  # Adjust spacing between subplots
plt.show()



# Image Display (if you have an image file named 'image.png' in the same directory)
try:  # Using try-except to handle potential FileNotFoundError
    img = plt.imread('image.png')
    plt.imshow(img)
    plt.title("Image")
    plt.axis('off')  # Hide axes
    plt.show()
except FileNotFoundError:
    print("image.png not found.  Skipping image display example.")



# 3D Plotting (requires mpl_toolkits.mplot3d)
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)  # Create a meshgrid
Z = np.sin(np.sqrt(X**2 + Y**2)) / np.sqrt(X**2 + Y**2)  # Example 3D function

ax.plot_surface(X, Y, Z, cmap='viridis') #cmap changes the color scheme
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
plt.title("3D Surface Plot")
plt.show()
```

```python
import pandas as pd
import numpy as np

# Creating a DataFrame

# From a dictionary
data = {'Name': ['Alice', 'Bob', 'Charlie', 'David'],
        'Age': [25, 30, 28, 22],
        'City': ['New York', 'London', 'Paris', 'Tokyo']}
df = pd.DataFrame(data)
print("DataFrame from Dictionary:\n", df)

# From a list of lists
data_list = [['Alice', 25, 'New York'],
             ['Bob', 30, 'London'],
             ['Charlie', 28, 'Paris'],
             ['David', 22, 'Tokyo']]
df_list = pd.DataFrame(data_list, columns=['Name', 'Age', 'City'])  # Specify column names
print("\nDataFrame from List of Lists:\n", df_list)

# From a NumPy array
data_np = np.random.randint(18, 35, size=(4, 2))  # 4 rows, 2 columns, ages between 18 and 34
df_np = pd.DataFrame(data_np, columns=['Age1', 'Age2'])
print("\nDataFrame from NumPy Array:\n", df_np)


# Basic DataFrame Operations

# Accessing columns
print("\nName Column:\n", df['Name'])
print("\nAge Column:\n", df.Age)  # Another way to access a column

# Accessing rows (using .loc and .iloc)
print("\nFirst Row:\n", df.loc[0])  # By label (index)
print("\nSecond Row:\n", df.iloc[1]) # By integer position

# Selecting data (slicing)
print("\nFirst Two Rows:\n", df.iloc[:2]) # First two rows
print("\nBob's Data:\n", df[df['Name'] == 'Bob']) # Rows where 'Name' is 'Bob'

# Adding a new column
df['Salary'] = [60000, 75000, 70000, 55000]
print("\nDataFrame with Salary Column:\n", df)

# Removing a column
df = df.drop('City', axis=1)  # axis=1 for columns
print("\nDataFrame after Removing City Column:\n", df)

# Renaming a column
df = df.rename(columns={'Salary': 'Annual Salary'})
print("\nDataFrame after Renaming Salary Column:\n", df)

# Descriptive Statistics
print("\nDescriptive Statistics:\n", df.describe())

# Sorting
print("\nSorted by Age:\n", df.sort_values(by='Age'))

# Grouping
print("\nGrouped by Age (mean salary):\n", df.groupby('Age')['Annual Salary'].mean())

# Handling Missing Data

# Creating a DataFrame with NaN values
data_nan = {'A': [1, 2, np.nan, 4], 'B': [5, np.nan, 7, 8], 'C': [9, 10, 11, np.nan]}
df_nan = pd.DataFrame(data_nan)
print("\nDataFrame with NaN values:\n", df_nan)

# Filling NaN values
df_filled = df_nan.fillna(0)  # Fill with 0
print("\nDataFrame after filling NaN with 0:\n", df_filled)

df_filled_mean = df_nan.fillna(df_nan.mean())  # Fill with mean of each column
print("\nDataFrame after filling NaN with mean:\n", df_filled_mean)

# Dropping rows with NaN values
df_dropped = df_nan.dropna()
print("\nDataFrame after dropping rows with NaN:\n", df_dropped)

# Reading and Writing Data (CSV)

# Sample data (replace with your file paths)
csv_file_path = "output.csv" # Example file path

# Writing to CSV
df.to_csv(csv_file_path, index=False)  # index=False to avoid writing row indices

# Reading from CSV
df_from_csv = pd.read_csv(csv_file_path)
print(f"\nDataFrame read from {csv_file_path}:\n", df_from_csv)


# More advanced operations (brief examples)

# Applying a function to a column
df['Age_Squared'] = df['Age'].apply(lambda x: x**2)
print("\nDataFrame with Age Squared column:\n", df)

# Merging DataFrames (similar to SQL JOIN)
df2 = pd.DataFrame({'Name': ['Alice', 'Bob'], 'Department': ['Sales', 'Marketing']})
merged_df = pd.merge(df, df2, on='Name')
print("\nMerged DataFrame:\n", merged_df)
```
